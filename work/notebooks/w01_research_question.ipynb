## 2. The question: decision, action, cost of a wrong call

Question: Which observable page characteristics (intent, content type,
position, freshness) are associated with a "Great Decoupling" signature —
impressions holding steady while clicks meaningfully decline over a recent
window — and can this be turned into an early-warning exposure signal?

Unit of analysis: one content page (content_id), evaluated using its
before/after 30-day windows.

Decision this improves: which pages an SEO/content strategist flags for
defensive review in a quarterly portfolio triage, before the click loss is
large enough to show up as an obvious revenue problem in a normal report.

Who acts, and how: a content or SEO strategist, using the flagged pages and
their reason codes to prioritize defensive work (e.g. adding original data,
restructuring for AI-citation capture, strengthening why a user should
still click) instead of routine "this page is old" refreshes.

Cost of a wrong call: false positive — the team spends scarce strategist
time defending a page whose click drop was really seasonality or a sibling
page absorbing demand, not decoupling. False negative — a real,
still-valuable page keeps silently losing clicks unnoticed, and the
eventual revenue loss is larger and harder to reverse than if it had been
caught early.

Why data/ML helps: this pattern is invisible to a normal glance at
impressions or rank alone — it only appears by deliberately comparing two
time windows against each other across thousands of pages, which is
exactly the kind of systematic check that's impractical by hand but
straightforward to compute and rank.

## 3. Quick look at the data (2-3 real numbers)
Loading data/raw/content_refresh_anonymized.csv to check whether AI-referral
patterns exist at all in this data, and how sparse they really are.

```python
import pandas as pd

df = pd.read_csv('data/raw/content_refresh_anonymized.csv')
n = len(df)

# only pages with real signal in both windows (avoids divide-by-zero noise)
sub = df[(df['impressions_prev_30d'] > 0) & (df['clicks_prev_30d'] > 0)].copy()

sub['impr_change_pct'] = 100 * (sub['impressions_last_30d'] - sub['impressions_prev_30d']) / sub['impressions_prev_30d']
sub['click_change_pct'] = 100 * (sub['clicks_last_30d'] - sub['clicks_prev_30d']) / sub['clicks_prev_30d']

# the signature: impressions roughly flat, clicks meaningfully down
pattern = sub[(sub['impr_change_pct'].between(-10, 10)) & (sub['click_change_pct'] <= -15)]

print(f"usable rows (real prior-window data): {len(sub)} ({100*len(sub)/n:.1f}% of {n})")
print(f"Great Decoupling-signature rows: {len(pattern)} ({100*len(pattern)/len(sub):.1f}% of usable rows)")
print(f"clients affected: {pattern['client_id'].nunique()} of {sub['client_id'].nunique()}")
print(f"median impressions_90d, affected vs all usable: {pattern['impressions_90d'].median()} vs {sub['impressions_90d'].median()}")
```

Result: 674 of 11,759 usable pages (5.7%) show the signature, spread
across 22 of 29 clients, with a median impressions_90d (4,232) nearly
identical to the portfolio-wide median (4,686) — meaning these are pages
with real, meaningful traffic, not dying long-tail pages. That's enough
evidence this pattern is real and worth building on.
